```python
import pandas as pd 
import numpy as np
```

```python
train_df = pd.read_csv("Training.csv")
test_df = pd.read_csv("Testing.csv")

df = pd.concat([train_df , test_df] , axis=0)
df.head(1)
```

```python
df = df.drop(["Unnamed: 133"] , axis=1)
```

```python
df.describe()
```

```python
df.isna().sum() 
```

```python
df = df.dropna() 
```

```python
df["prognosis"].unique().size
```

```python
print(df["prognosis"].unique())
```

```python
import re
def replace_space(x) : 
    return re.sub(r'\(.*?\)' , '' , x).strip().lstrip().rstrip().replace(" " , "_")
df["prognosis"] = df["prognosis"].apply(replace_space)
print(df["prognosis"].unique())
```

```python
from sklearn.feature_extraction.text import CountVectorizer 

vectorizer = CountVectorizer()
result = vectorizer.fit_transform(df["prognosis"]) 
print(vectorizer.get_feature_names_out())
labels = pd.DataFrame( result.toarray() , columns=vectorizer.get_feature_names_out() )
labels.head()
```

```python
labels.shape
```

```python
df = df.drop(["prognosis"] , axis = 1 )
```

```python
from sklearn.model_selection import train_test_split 

x_train , x_test , y_train , y_test = train_test_split(df , labels , test_size=0.1 , random_state=0)
```

```python
print("the shape of x_train is" , x_train.shape) 
print("the shape of y_train is" , y_train.shape) 
print("the shape of x_test is" , x_test.shape) 
print("the shape of y_test is" , y_test.shape) 
```

```python
print(y_test.values)
```

```python
from keras.models import Sequential
from keras.layers import Dense , Input , Dropout
from keras.optimizers import SGD , Adam

model = Sequential()
model.add(Input(shape=(x_train.shape[1],)))
model.add(Dense(150, activation="relu"))
model.add(Dropout(0.3))  # 30% of the neurons will be randomly dropped
model.add(Dense(100, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(90, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(y_train.shape[1], activation="sigmoid"))

model.summary()

# Compile the Model
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Change optimizer to Adam and adjust learning rate
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
model.fit(x_train.values 
, y_train.values 
, epochs=100 
, batch_size=32
, validation_split=0.2)
```

```python
model.evaluate(x_test.values , y_test.values)
```

```python
res = model.predict(x_test.values)
print("data" , x_test.values[0])
print("result" , vectorizer.get_feature_names_out()[res[0].argmax()])
```

```python
new_row = {col : 0 for col in x_train.columns } 
print(new_row)
for i , value in enumerate( x_test.values[0] ):
    if value == 1 :
        new_row[x_train.columns[i]] = 1 
```

```python
model.save("disease_model.h5")
```
